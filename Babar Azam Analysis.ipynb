```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv("babar.csv")
data
```

```python
data.info()
```

### Data is already clean so lets start analysis

## First we check max Runs against each team

```python
data[["Opposition","Runs"]]
```

```python
Runs = data[["Opposition","Runs"]].sort_values(by="Runs",ascending = False)
Runs
```

```python
plt.bar(Runs["Opposition"],Runs["Runs"])
plt.xticks(rotation = 90)
plt.show()
```

### Hypothesis
##### Highest score is against South Africa

## Checking Max 6s against each team

```python
six = data[["Opposition","6s"]].sort_values(by="6s",ascending = False)
six
```

```python
plt.bar(six["Opposition"],six["6s"])
plt.xticks(rotation = 90)
plt.show()
```

### Hypothesis
#### Max 6s is against South Africa

## Checking Highest strike rate

```python
sr = data[["Opposition","SR"]].sort_values(by="SR", ascending = False)
sr
```

```python
plt.bar(sr["Opposition"],sr["SR"])
plt.xticks(rotation = 90)
plt.show()
```

### Hypothesis
#### Max SR is against South Africa
